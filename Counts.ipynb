```python
from math import floor, log
```

```python
def count_solutions(base, k):
    count = 0
    I = (base ** k - 1) // (base - 1) - 1
    for b in range(2, base):
        for c_k in range(1, b):
            c = b * I + c_k
            a = b * c // (b * base - (base - 1) * c_k)
            if a < base ** (k - 1):
                continue
            #if a % base == b:
            #    continue
            if (a * base + b) * c == a * (b * base**k + c):
                count += 1
    return count
```

```python
counts = dict()
```

```python
for base in range(2, 10000):
    if base in counts:
        continue
    k = 2 * floor(log(base, 2)) + 1
    counts[base] = count_solutions(base, k)
    print(base, counts[base])
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['figure.figsize'] = [10, 10]

bases = np.array(list(counts.keys()))
y     = np.array(list(counts.values()))

plt.scatter(bases, y / bases)
```
